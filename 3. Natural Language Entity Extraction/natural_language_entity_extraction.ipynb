```python
import pandas as pd
import spacy
```

```python
df = pd.read_csv("final_cbc_diagnoses_dataset_with_labels.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_entities(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities
```

```python
df['short_title_entities'] = df['short_title'].apply(lambda x: extract_entities(str(x)))
df['long_title_entities'] = df['long_title'].apply(lambda x: extract_entities(str(x)))
```
